## Creating a table named "Doctors" with "DocIDN" as the primary key

```
df_3 = _dntk.execute_sql(
  'drop table if exists Doctors;\ncreate table Doctors(\nDocLName varchar(500) NOT NULL, \nDocIDN int NOT NULL PRIMARY KEY,  \nSpecialty varchar(20) not Null);\ndescribe Doctors;',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_3
```

## Creating a table named "Patients"

```
df_5 = _dntk.execute_sql(
  'drop table if exists Patients;\ncreate table Patients(\nFK_DocIDN int NOT NULL, \nPatientStatus varchar(20) not null, \nPatientBirthday date not null,\nForeign Key (FK_DocIDN) references Doctors(DocIDN)); \ndescribe Patients;',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_5
```

## Creating a table named "Salary" for the doctor's salaries and their home city

## Finding total salary using DocIDN

```
df_11 = _dntk.execute_sql(
  'SELECT DocIDN, SUM(MonthlySalary) as TotalSalary\nFROM df_8\nWHERE DocIDN = 123 OR DocIDN = 456\nGROUP BY DocIDN;',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_11
```

## Finding speciality of a doctor using name

```
df_13 = _dntk.execute_sql(
  'SELECT \n    CASE \n        WHEN DocLName = \'Howard\' THEN CONCAT(\'Speciality of \', DocLName, \' is \', Specialty)\n        WHEN DocLName = \'Gravot\' THEN CONCAT(DocLName, \' is not a doctor\')\n    END AS Specialty_of_Doc\nFROM df_4\nWHERE DocLName IN (\'Howard\', \'Gravot\');',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_13
```

## Updating salary of the doctors

```
df_14 = _dntk.execute_sql(
  'UPDATE Salary \nSET MonthlySalary = MonthlySalary + 10 \nWHERE MonthlySalary < 26000;\nSELECT * FROM Salary;',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_14
```

## Selecting and retrieving patient info based on the condition of the patient

```
df_15 = _dntk.execute_sql(
  'SELECT \n    CASE \n        WHEN PatientStatus = \'Recovering\' THEN CONCAT(\'Birthday of recovering patient is \', PatientBirthday)\n        WHEN PatientStatus = \'Critical\' THEN CONCAT(\'Birthday of critical patient is \', PatientBirthday)\n        ELSE CONCAT(PatientStatus, \' is not a patient\')\n    END AS Patient_Info\nFROM Patients\nWHERE PatientStatus IN (\'Recovering\', \'Critical\');',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_15
```

## Getting average salary

```
df_16 = _dntk.execute_sql(
  'SELECT AVG(MonthlySalary) as AvgSalary FROM Salary;',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_16
```

## Getting monthly salary by name

```
df_17 = _dntk.execute_sql(
  'SELECT \n    CASE \n        WHEN DocLName = \'Matthews\' THEN CONCAT(\'Monthly salary of \', DocLName, \' is $\', MonthlySalary)\n        ELSE CONCAT(\'Matthews\', \' is not a Doctor\')\n    END AS DocSalary\nFROM df_10\nWHERE DocLName = \'Matthews\';',
  'SQL_DEEPNOTE_DATAFRAME_SQL',
  audit_sql_comment='',
  sql_cache_mode='cache_disabled',
  return_variable_type='dataframe'
)
df_17
```
